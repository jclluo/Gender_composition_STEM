```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup
```

```python
ucichagowebsite = "https://economics.uchicago.edu/people/job-market-candidates"
response = requests.get(ucichagowebsite)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
chicago_content = soup.text
chicago_content
```

```python
chicago_substring = chicago_content[3720:10039]
candidates = chicago_substring.split('\n\n')
candidates
```

```python
candidates = chicago_substring.split('\n\n')

data = []

for i, candidate in enumerate(candidates):
    if i % 2 == 0:
        details = candidate.split('\n')
        
        name = details[0].strip()
        primary_focus = details[1].replace('Primary Research Focus: ', '').strip()
        secondary_focus = details[2].replace('Secondary Research Focus: ', '').strip()
        references = details[3].replace('References: ', '').strip()
        
        email = details[5].replace('Email: ', '').strip()
        
        website_text = details[6].replace('Personal Website: ', '').strip()
        website_tag = None
        for a_tag in soup.find_all('a'):
            if website_text in a_tag.text:
                website_tag = a_tag
                break
        website = website_tag.text.strip() if website_tag else ''
        website_link = website_tag['href'] if website_tag else ''
        
        data.append({
            'Name': name,
            'Primary Research Focus': primary_focus,
            'Secondary Research Focus': secondary_focus,
            'References': references,
            'Email': email,
            'Website': website,
            'Website Link': website_link
        })

uchicago_2023_2024 = pd.DataFrame(data)

uchicago_2023_2024
```

```python
candidate_blocks = soup.find_all('div', class_='paragraph paragraph--type--uc-text-block paragraph--view-mode--default')

job_market_data = []

for block in candidate_blocks:
    name_tag = block.find('h2')
    if name_tag and name_tag.find('a'):
        name = name_tag.get_text(strip=True)

    job_market_paper = ""
    job_market_paper_link = ""

    job_market_paper_tag = block.find(string=lambda text: "Job Market Paper Title" in text)
    
    if job_market_paper_tag:
        a_tag = job_market_paper_tag.find_next('a')
        if a_tag:
            job_market_paper = a_tag.get_text(strip=True)
            job_market_paper_link = a_tag['href']

    if job_market_paper and job_market_paper_link:
        job_market_data.append({
            'Name': name,
            'Job Market Paper': job_market_paper,
            'Job Market Paper Link': job_market_paper_link
        })

website_chicago = pd.DataFrame(job_market_data)
website_chicago
```

```python
chicago_candidate_23_24 = uchicago_2023_2024.merge(website_chicago, on='Name',how='left')
chicago_candidate_23_24.iloc[0, chicago_candidate_23_24.columns.get_loc('Job Market Paper')] = 'Sticks vs Carrots: Optimal Climate Policy with Government Turnovers'
chicago_candidate_23_24.iloc[0, chicago_candidate_23_24.columns.get_loc('Job Market Paper Link')] = 'https://socialsciences.uchicago.edu/sites/default/files/2023-11/JMP_Behmer_1.pdf'

chicago_candidate_23_24
```

```python
chicago_candidate_23_24.to_csv('chicago_candidate_23_24.csv', index=False)
```

```python

url = "https://econ.berkeley.edu/graduateplacement/job-market-candidates-phd"

response = requests.get(url)

soup_2 = BeautifulSoup(response.content, 'html.parser')

candidates_section = soup_2.find('div', class_='view-content')
with open('candidates_section.html', 'w', encoding='utf-8') as file:
    file.write(str(candidates_section))

```

```python
candidate_blocks = soup_2.find_all('div', class_='teaser_display')

ucb_candidates_data = []

for block in candidate_blocks:
    entry_year_tag = block.find('div', class_='display-program-entry')
    entry_year = entry_year_tag.get_text(strip=True) if entry_year_tag else None

    name_tag = block.find('div', class_='display-name')
    name = name_tag.get_text(strip=True) if name_tag else None

    fields_tag = block.find('div', class_='display-fields')
    fields = fields_tag.get_text(strip=True) if fields_tag else None

    homepage_tag = block.find('div', class_='display-homepage')
    homepage_link = homepage_tag.find('a')['href'] if homepage_tag and homepage_tag.find('a') else None

    email_tag = block.find('div', class_='display-email')
    email_link = email_tag.find('a')['href'].replace('mailto:', '') if email_tag and email_tag.find('a') else None

    ucb_candidates_data.append({
        'Name': name,
        'Entry Year': entry_year,
        'Fields': fields,
        'Website': homepage_link,
        'Email': email_link
    })

df_ucb_2024_2025 = pd.DataFrame(ucb_candidates_data)
df_ucb_2024_2025.to_csv('ucb_2024_2025.csv', index=False)
df_ucb_2024_2025['School'] = 'UC Berkeley'
df_ucb_2024_2025['Year'] = '2024-2025'
df_ucb_2024_2025
```
